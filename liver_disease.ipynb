```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
liver_df=pd.read_csv('indian_liver_patient.csv')
liver_df.head()
```

```python
liver_df.info()
```

```python
liver_df.isnull().sum()
```

```python
liver_df.fillna(method='ffill',inplace=True)
```

```python
liver_df.isnull().sum()
```

```python
liver_df.duplicated()
```

```python
liver_df['Gender'].replace({'Male':1,'Female':0},inplace=True)
```

```python
liver_df['Dataset'].replace({1:0,2:1},inplace=True)
```

```python
liver_df
```

```python
liver_df.corr()
```

```python
x=liver_df.drop('Dataset',axis=1)
y=liver_df['Dataset']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier().fit(x_train,y_train)
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
model1=XGBClassifier().fit(x_train,y_train)
y_pred=model1.predict(x_test)
print(classification_report(y_test,y_pred))
```
